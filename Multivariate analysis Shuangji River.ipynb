# **Factor Analysis**

Factor analysis is a statistical method used to identify the relation between variables and can be used to reduct the dimension of data. The aim is to get variables that can explain the variation among data the most. Besides that, the dimension reduction can make the data structure is easier to learn.

This method was chosen to reduce the dimension and to considerate the correlation pattern between variables when PCA (Principal Component Analysis) is only focusing on total variance.

# Assumption Test
**Covariance Matrix Homogeneity Test -> Bartlett test**

Bartlett's test is used in Factor Analysis to test whether the covariance matrix between variables is homogeneous enough to carry out factor analysis. In other words, this test helps determine whether the variables to be used in factor analysis are related or not.

More technically, the Bartlett test tests the null hypothesis that the covariance matrix of these variables is an identity matrix, which means that these variables are not related to each other at all. If the p value (p-value) from the Bartlett test is smaller than the specified significance level (usually 0.05), then we can reject the null hypothesis, and this indicates that the covariance matrix between variables is not identical, so factor analysis can be carried out.

Conceptually, Factor Analysis requires the assumption that the observed variables have a latent structure that can be explained by these factors. Bartlett's test helps ensure that these assumptions are adequate to conduct factor analysis.

If the results of Bartlett's test indicate that the covariance matrix is â€‹â€‹not identical (p value < 0.05), this supports the use of factor analysis to try to identify the factors underlying the variation in those variables. If the Bartlett test results are not supportive (p value > 0.05), then it may be better to look for a more appropriate analysis method or make modifications to the data before proceeding with factor analysis.

```python
from factor_analyzer.factor_analyzer import calculate_bartlett_sphericity
bartlett_test_statistic, bartlett_p_value = calculate_bartlett_sphericity(df_normalized)

# Show result
print(f"Bartlett Test Statistics: {bartlett_test_statistic}")
print(f"p-Value: {bartlett_p_value}")

# Interpretation
if bartlett_p_value < 0.05:
    print("Reject the null hypothesis. The covariance matrix is not the same for at least one pair of variables.")
else:
    print("Fail to reject the null hypothesis. Covariance matrices are the same for all variables.")
```

**KMO Test**

KMO Test (Kaiser-Meyer-Olkin) is a test to evaluate if the data is sufficient or not for factor analysis. This test can show how far the data can be explained by latent factors or non-observation factors. If the value is close to 1, the correlation between variables is pretty strong, indicates that the data is sufficient.

```python
from factor_analyzer.factor_analyzer import calculate_kmo

kmo, kmo_model = calculate_kmo(df_normalized)
kmo_pervar = pd.DataFrame(kmo, index = list(df_normalized.columns),
                         columns=['Nilai KMO']) #KMO for each variable
print(kmo_pervar)
print(f"KMO Model   : {kmo_model}")

# Evaluation if the data is sufficient for factor analysis
if kmo_model > 0.6:
    print("Data is sufficient for factor analysis")
else:
    print("Data is insufficient for factor analysis")
```

```python
from factor_analyzer import FactorAnalyzer
import pandas as pd

# Function to get eigenvalues, variance proportion, and cumulative variance proportion
def get_eigenvalues(model):
    eigenvalues = model.get_eigenvalues()
    total_variance = sum(eigenvalues[0])
    variance_proportion = eigenvalues[0] / total_variance
    cumulative_variance = variance_proportion.cumsum()

    eigenvalues_df = pd.DataFrame({
        'Factor': range(1, len(eigenvalues[0]) + 1),
        'Eigenvalues': eigenvalues[0],
        'Proportion Variance': variance_proportion * 100,
        'Cumulative Variance': cumulative_variance * 100
    })
    return eigenvalues_df

# Oblimin Rotation
model_fa_oblimin = factor_model(df_selected_filtered, number_of_factors=len(df_selected_filtered.columns), rotation='oblimin')

eigenvalues_df = get_eigenvalues(model_fa_oblimin)
print(eigenvalues_df)



```

We use eigenvalue = 1 and cumulative proportion = 60% as a threshold

```python
import matplotlib.pyplot as plt

# Plot scree plot
plt.figure(figsize=(10, 6))
plt.plot(eigenvalues_df['Factor'], eigenvalues_df['Eigenvalues'], marker='o', linestyle='-', color='b', label='Eigenvalues')
plt.title('Scree Plot')
plt.xlabel('Faktor')
plt.ylabel('Eigenvalues')
plt.axhline(y=1, color='red', linestyle='--', label='Threshold Eigenvalues=1')  # Tambahkan garis pada Eigenvalues=1 dengan warna merah
plt.legend()

plt.grid(which='both', linestyle='--', linewidth=0.5, color='gray', alpha=0.3)  # Atur alpha untuk membuat grid transparan

plt.show()
```

```python
import matplotlib.pyplot as plt

# Plot variance
plt.figure(figsize=(10, 6))
plt.plot(eigenvalues_df['Factor'], eigenvalues_df['Proportion Variance'], marker='o', linestyle='-', color='b', label='Proporsi Variance')
plt.plot( eigenvalues_df['Factor'], eigenvalues_df['Cumulative Variance'],marker='o', linestyle='-', color='r', label='Kumulatif Proporsi Variance')
plt.title('Variance')
plt.xlabel('Nomor Komponen')
plt.ylabel('Proporsi Variance')
plt.xticks(range(1, len(eigenvalues_df['Factor']) + 1))  # Set interval sumbu x menjadi 1
plt.yticks(range(0, 101, 10))  # Set interval sumbu y menjadi 10
plt.axhline(60, color='red', linestyle='--', label='Threshold 60%')  # Tambahkan garis pada 70% dengan warna merah
plt.legend()

# Hilangkan grid di belakang plot
plt.grid(which='both', linestyle='--', linewidth=0.5, color='gray', alpha=0.3)  # Atur alpha untuk membuat grid transparan

plt.show()
```

Grouping variables into factors is based on the largest factor loading value or the absolute value of the largest factor loading.

This value indicates the extent to which a variable influences or is associated with a particular factor in factor analysis.

```python
# Dominant factors for each variable based on the biggest absolute of loading value
dominant_factors = loading_df.abs().idxmax(axis=1)

faktor_variabel_df = pd.DataFrame(index=[f'Faktor_{i+1}' for i in range(factors_number)], columns=['Variabel'])

# Iteration through each factor
for faktor in faktor_variabel_df.index:
    variabel_terkait = dominant_factors[dominant_factors == faktor].index.tolist()
    faktor_variabel_df.at[faktor, 'Variabel'] = variabel_terkait

print(faktor_variabel_df)
```

# **Spatial Clustering**

In this case, we were using Hierarchial Clustering. This method was chosen because the little number of spatial data (14 points of monitoring) and temporal data (12 months). Furthermore, this method can produce dendrogram that can visualized the relationship of each group. This method is also suitable when the data do not have any information about the amount of clusters.

```python
# Applying factor model into extracted data
transformed_data = model_fa_selected_factors.transform(df_selected_filtered)

# Make a dataframe from the transformation result
df_transformed = pd.DataFrame(transformed_data, columns=[f'Factor_{i+1}' for i in range(factors_number)])

print(df_transformed)
```

```python
#add spatial point information into a column
dataspasial = pd.concat([df_transformed, df["NO."]], axis = 1)
dataspasial.head()
```

```python
#to calculate the factor mean in each place
dataspasial_grouped = dataspasial.groupby("NO.").mean().reset_index()
dataspasial_grouped
```

```python
from scipy.spatial.distance import pdist, squareform
from scipy.cluster.hierarchy import linkage, dendrogram
import matplotlib.pyplot as plt
```

Ward's minimum method is an approach in agglomerative clustering where the methos is minimize he increase of variance of sum of square of the differences between observations in a cluster, everytime two clusters are combined.

```python
#calculate the pairwise distances between the rows in dataframe
sq_distance = pdist(dataspasial_grouped[['Factor_1', 'Factor_2', 'Factor_3', 'Factor_4']])
#squareform, convert the condensed dist matrix into a full square-form distance matrix
sq_distance = squareform(sq_distance)

#hierarchical clustering, each step combines two clusters that minimize the variance within clusters
linkage_matrix = linkage(sq_distance, method = 'ward')

# Creating the dendrogram
dendrogram(linkage_matrix, labels=[str(label) for label in dataspasial_grouped['NO.'].tolist()])

# Adjusting the font size of the labels
plt.tick_params(axis='x', labelsize=11)
plt.xlabel('Data Points', fontsize=12)
plt.ylabel('Euclidean Distance', fontsize=12)

# Adding title
plt.title('Dendrogram Spasial', fontsize=14)

# Displaying the plot
plt.show()
```

The longer the line between clusters, the more different the clusters are, and the shorter the line, the more similar the clusters are. Therefore, can be concluded from the dendrogram, the selected clusters are 3 clusters.

```python
from sklearn.cluster import KMeans
from sklearn.metrics import silhouette_score
import matplotlib.pyplot as plt
```

# Elbow and Silhouette Method to Choose the Number of Clusters

```python
inertias = []
for k in range(1, 14):
    kmeans = KMeans(n_clusters=k)
    kmeans.fit(dataspasial_grouped[['Factor_1', 'Factor_2', 'Factor_3', 'Factor_4']])
    inertias.append(kmeans.inertia_)

plt.figure(figsize=(8, 5))
plt.plot(range(1, 14), inertias, marker='o')
plt.title('Metode Elbow')
plt.xlabel('Jumlah Cluster')
plt.ylabel('Inertia')
plt.show()
```

```python
#Penentuan K cluster metode Silhouette
silhouette_scores = []
for k in range(2, 14):
    kmeans = KMeans(n_clusters=k)
    kmeans.fit(dataspasial_grouped[['Factor_1', 'Factor_2', 'Factor_3', 'Factor_4']])
    score = silhouette_score(dataspasial_grouped[['Factor_1', 'Factor_2', 'Factor_3', 'Factor_4']], kmeans.labels_)
    silhouette_scores.append(score)

# Plotting grafik Silhouette
plt.figure(figsize=(8, 5))
plt.plot(range(2, 14), silhouette_scores, marker='o')
plt.title('Metode Silhouette')
plt.xlabel('Jumlah Cluster')
plt.ylabel('Silhouette Score')
plt.show()
```

From the Elbow Method, the optimum k value at the point at the elbow is k=3. For the silhouette method, the optimum k is obtained when the silhouette score value is highest, namely k=3. Therefore, many clusters, k = 3.

```python
from sklearn.cluster import AgglomerativeClustering

clusters = AgglomerativeClustering(n_clusters=3, metric='euclidean', linkage='ward')
cluster_labels = clusters.fit_predict(dataspasial_grouped[['Factor_1', 'Factor_2', 'Factor_3', 'Factor_4']])

print(cluster_labels)
```

```python
cluster_spasial = pd.Series(clusters.labels_, name = 'Cluster Spasial')
df_spasial=pd.concat([dataspasial_grouped, cluster_spasial], axis = 1)
print(df_spasial)
```

```python
groupedby_spatial = pd.concat([dataspasial_grouped, cluster_spasial], axis=1)
grouped = groupedby_spatial.groupby('Cluster Spasial')
result = grouped.mean(numeric_only=True)
print(result)
```
